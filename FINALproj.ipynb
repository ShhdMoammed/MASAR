## 1- Data cleaning

````python
url = '/content/Used Car Dataset.csv'
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

#Read data from CSV file
df = pd.read_csv(url)
df
````

````python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Correlation Heatmap')
plt.show()
````

## 2- Data processing

### normalization

````python
df.dtypes
````

````python
scaler = StandardScaler()
cols_to_normalize = ['mileage(kmpl)','engine(cc)','max_power(bhp)','torque(Nm)','price(in lakhs)']
scaled_data = scaler.fit(df[cols_to_normalize])
df[cols_to_normalize] = scaler.transform(df[cols_to_normalize])
min_max_scaler = MinMaxScaler()
df[cols_to_normalize] = min_max_scaler.fit_transform(df[cols_to_normalize])
````

1 . Loop: Question: Write a Python program that uses a 'for' loop to print all even numbers from 1 to 10. Then, use 'while' loop.

````python
# Using for loop
for num in range(1, 11):
    if num % 2 == 0:
        print(num)

# Using while loop
num = 1
while num <= 10:
    if num % 2 == 0:
        print(num)
    num += 1
````

2 .Control Flows: Question: Create a Python function named grade_classifier that takes a student's score as an argument and returns a string indicating their grade. Use the following criteria:
Score >= 90: "A" 80 <= Score < 90: "B" 70 <= Score < 80: "C" 60 <= Score < 70: "D" Score < 60: "F"

````python
def grade_classifier(score):
    if score >= 90:
        return "A"
    elif 80 <= score < 90:
        return "B"
    elif 70 <= score < 80:
        return "C"
    elif 60 <= score < 70:
        return "D"
    else:
        return "F"
````
